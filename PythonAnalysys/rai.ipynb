```python
from scipy.interpolate import interp1d
import numpy as np

xs = np.arange(10)
ys = 2*xs + 1

interp_func = interp1d(xs, ys)

newarr = interp_func(np.arange(2.1, 3, 0.1))

print(newarr)
reverse_newar = list(reversed(newarr))
print(reverse_newar)
```

```python
import matplotlib.pyplot as plt
plt.plot(newarr, marker='>', linestyle='', color='b', markersize=5)  # 'o' означает точки, 'linestyle=''' означает отсутствие линий
plt.plot(reverse_newar, marker='<', linestyle='', color='green', markersize=5)  # 'o' означает точки, 'linestyle=''' означает отсутствие линий
# Добавляем названия осей
plt.xlabel('X')
plt.ylabel('Y')

# Добавляем заголовок
plt.title('График точек')

# Добавляем сетку
plt.grid(True)

# Отображаем график
plt.show()
```
